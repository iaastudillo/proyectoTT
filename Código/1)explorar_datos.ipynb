## Función para dibujar una gráfica de pastel

```python
def dibujarPolaridadPie():
    df.POLARIDAD.value_counts().plot(
        kind='pie', 
        autopct='%1.0f%%',
        startangle=110, 
        explode = (0.1, 0.1, 0.1),
        figsize = (7,7),
        fontsize = 12,
        colors = ['#D68910', '#27AE60', '#5DADE2'],
        title = "Frecuencia por número de observaciones")
```

## Función para dibujar una gráfica de barras

```python
def dibujarPolaridadBar():
    df.POLARIDAD.value_counts().plot(
        kind='bar',
        figsize=(7,7),
        fontsize = 14,
        rot=0,
        title = "Frecuencia por número de observaciones",
        xlabel = "Clase",
        color = ['#D68910', '#27AE60', '#5DADE2'],
        edgecolor='black',
        ylabel = "Número de observaciones")
```

```python
dibujarPolaridadPie()
```

```python
dibujarPolaridadBar()
```
